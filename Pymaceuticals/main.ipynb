## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use this straighforward method, create multiple series and put them all in a dataframe at the end. ['Tumor Volume (mm3)']
meandata = dropped_df[['Drug Regimen','Tumor Volume (mm3)']].rename(columns={"Tumor Volume (mm3)": "Average Tumor Volume"})
meandata
mean_df = meandata.groupby(['Drug Regimen']).mean()
# mean_df
mediandata = dropped_df[['Drug Regimen','Tumor Volume (mm3)']].rename(columns={"Tumor Volume (mm3)": "Median Tumor Volume"})
median_df = mediandata.groupby(['Drug Regimen']).median()
# median_df
vardata = dropped_df[['Drug Regimen','Tumor Volume (mm3)']].rename(columns={"Tumor Volume (mm3)": "Variance of Tumor Volume"})
var_df = vardata.groupby(['Drug Regimen']).var()
# var_df
stddata = dropped_df[['Drug Regimen','Tumor Volume (mm3)']].rename(columns={"Tumor Volume (mm3)": "StdDev Tumor Volume"})
std_df = stddata.groupby(['Drug Regimen']).std()
# std_df
semdata = dropped_df[['Drug Regimen','Tumor Volume (mm3)']].rename(columns={"Tumor Volume (mm3)": "StdErr Tumor Volume"})
sem_df = semdata.groupby(['Drug Regimen']).sem()
# sem_df
# median_df = dropped_df.groupby(['Drug Regimen']).median()
# var_df = dropped_df.groupby(['Drug Regimen']).var()
# std_df = dropped_df.groupby(['Drug Regimen']).std()
# sem_df = dropped_df.groupby(['Drug Regimen']).sem()
summarystats = pd.merge(mean_df, median_df, on="Drug Regimen")
summarystats = pd.merge(summarystats, var_df, on="Drug Regimen")
summarystats = pd.merge(summarystats, std_df, on="Drug Regimen")
summarystats = pd.merge(summarystats, sem_df, on="Drug Regimen")
# , , std_df, sem_df,
# summarystats = pd.DataFrame([mean_df, median_df, var_df, std_df, sem_df])
summarystats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use method to produce everything with a single groupby function
summarydata = dropped_df.groupby(['Drug Regimen'])
summarydata.head()
summarymean = summarydata["Tumor Volume (mm3)"].mean()
summarymedian = summarydata["Tumor Volume (mm3)"].median()
summaryvar = summarydata["Tumor Volume (mm3)"].var()
summarystd = summarydata["Tumor Volume (mm3)"].std()
summarysem = summarydata["Tumor Volume (mm3)"].sem()
summarystats2 = pd.DataFrame({"Average Tumor Volume": summarymean, 
                              "Median Tumor Volume":summarymedian, 
                              "Variance of Tumor Volume":summaryvar, 
                              "StdDev Tumor Volume": summarystd, 
                              "StdErr Tumor Volume": summarysem})
summarystats2
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
from scipy.stats import linregress

x_values = scatterdata['Weight (g)']
y_values = scatterdata['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Average Tumor Volume per Weight for Capomulin')
plt.show()
```
